```python
# Rename
train_df = train_df.rename(columns={'Province/State':'Province_state','Country/Region':'Country_Region','Lat':'Latitude','Long':'Longitude','ConfirmedCases':'Confirmed_cases'})
```

```python
train_df.columns
```

```python
# count number of null data in dataframe
train_df.isnull().sum()
```

```python
# fill data with country
train_df['Province_state'] = train_df['Province_state'].fillna(train_df['Country_Region'])
```

```python
train_df.head()
```

```python
train_df.drop('Date_new', axis =1,inplace = True) # drop date -> column drop
```

```python
train_df.head()
```

```python
# - replace
train_df['Date'] = train_df['Date'].apply(lambda x :x.replace('-','')) # str -> int
# date column as integer
train_df['Date'] = train_df['Date'].astype(int)
```

```python
train_df.head()
```

```python
test_df["Date"]=test_df["Date"].apply(lambda x :x.replace("-",""))
                                                          
```

```python
test_df["Date"]=test_df["Date"].astype(int)
```

```python
test_df.head()
```

### process for training

```python
X_train = train_df[['Latitude','Longitude','Date']] # features # first csv
```

```python
y1 = train_df[['Confirmed_cases']] # targets # first csv
y2 = train_df[['Fatalities']]
```

```python
X_test = test_df[['Lat','Long','Date']] # second csv
```

```python
# Random forest model
from sklearn.ensemble import RandomForestClassifier
```

```python
tree_model = RandomForestClassifier(max_depth = 200, random_state=0)
```

```python
# Train
tree_model.fit(X_train,y1)
```

```python
pred1 = tree_model.predict(X_test)
pred1 = pd.DataFrame(pred1)
pred1.columns = ['confirm_cas_prediction']
```

```python
tree_model.fit(X_train,y2)
```

```python
pred2 = tree_model.predict(X_test)
pred2 = pd.DataFrame(pred2)
pred2.columns = ['Fatalities_prediction']
```
